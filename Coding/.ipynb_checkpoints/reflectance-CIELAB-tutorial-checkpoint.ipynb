```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lam = np.linspace(400, 700, 31)
R = 0.1*((0.7-0.1)/(560-480))*(lam > 480)*(lam < 560) + ((0.8-0.7)/(700-560))*(lam >= 560)*(lam <= 700)
```

```python
plt.plot(lam, R, 'yx-')
plt.xlabel('wavelength $\lambda$ (nm)')
plt.ylabel('Reflectance $R(\lambda)$');
```
